```python
import json

# Load JSON
with open("datasets/Wikilingual-dataset/train.json", encoding="utf-8") as f:
    data = json.load(f)

# Convert dict → list of (key, value)
items = list(data.items())

# Lấy 2000 mẫu đầu
items_2k = items[:2000]

# Convert lại thành dict
data_2k = dict(items_2k)

# Save file mới
with open("datasets/Wikilingual-dataset/train_2k.json", "w", encoding="utf-8") as f:
    json.dump(data_2k, f, ensure_ascii=False, indent=4)
print("Đã tạo file train_2k.json với 2000 mẫu!")
```
